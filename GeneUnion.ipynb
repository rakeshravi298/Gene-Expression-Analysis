```python
import GEOparse

def FullFunction(s):

    """
    Run FullFunction
    Input : Give the series number of your experiment Example: GSE63514
    Output : Returns the dataframe where it combines all the sanple's data for each and every gene
    """
    
        
    df = pd.DataFrame()
    str = ""
    gse = GEOparse.get_GEO(geo=s, destdir="./")
    
    for key in gse.gsms.keys():
        str = key
        break
        
    lst = gse.gsms[str].table["ID_REF"]

    i=0
    dict = {}
    for k in lst:
        dict[i] = k
        i=i+1
        
    for key in gse.gsms.keys(): #gets the key value that is the sample
        
        df = df._append(gse.gsms[key].table["VALUE"])

    df.rename(columns = dict, inplace = True)
    df.insert(0, "SampleId", [*gse.gsms.keys()], True)

    target_lst = []
    
    for key in gse.gsms.keys():
        if(s=="GSE19188"):
            target_lst.append(gse.gsms[key].metadata["characteristics_ch1"][0])
        else:
            target_lst.append(gse.gsms[key].metadata["title"][0])
    
    df.insert(1, "Target", target_lst, True)
    
    return df
```

```python
GSE42568, GSE19188, GSE121248, 
GSE63514, and GSE66272
```

```python
df_1 = FullFunction("GSE42568")
df_2 = FullFunction("GSE19188")
df_3 = FullFunction("GSE121248")
df_4 = FullFunction("GSE63514")
df_5 = FullFunction("GSE66272")
```

```python
df_1
```

```python
df_merged = pd.concat([df_1, df_2, df_3 , df_4 , df_5], ignore_index=True, sort=False)
```

```python
import re

# text = "adj.normal kidney sample, patient 13_K851_NN"

# # Search for the word "normal" (case-insensitive)
# if re.search(r'\bnormal\b', text, re.IGNORECASE):
#     print("The word 'normal' is present in the string.")
# else:
#     print("The word 'normal' is not present in the string.")

def check_normal(s):
    if re.search(r'\bnormal\b', s, re.IGNORECASE):
        return "Normal"
    elif re.search(r'\bhealthy\b', s, re.IGNORECASE):
        return "Normal"
    elif re.search(r'\btumor\b', s, re.IGNORECASE):
        return "Cancer"
    elif re.search(r'\bcancer\b', s, re.IGNORECASE):
        return "Cancer"
    elif re.search(r'\bccrcc\b', s, re.IGNORECASE):
        return "Cancer"
    elif re.search(r'\bCIN1\b', s, re.IGNORECASE):
        return "Cancer"
    elif re.search(r'\bCIN2\b', s, re.IGNORECASE):
        return "Cancer"
    elif re.search(r'\bCIN3\b', s, re.IGNORECASE):
        return "Cancer"
    else:
        return s

df_merged["Target"] = df_merged["Target"].apply(check_normal)
```
